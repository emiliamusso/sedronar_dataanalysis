```python
import plotly.express as px

data = {
    "Sustancia": [
        "Heroína", "Cocaína", "Alcohol", "Ketamina", "Benzodiacepinas",
        "Anfetaminas", "Tabaco", "Marihuana", "Solventes", "LSD", "GHB", "Éxtasis", "Poppers"
    ],
    "Daño físico": [
        2.78, 2.33, 1.40, 2.00, 1.63, 1.81, 1.24, 0.99, 1.28, 1.13, 0.86, 1.05, 0.93
    ],
    "Dependencia": [
        3.00, 2.39, 1.93, 1.54, 1.83, 1.67, 2.21, 1.51, 1.01, 1.23, 1.19, 1.13, 0.87
    ],
    "Daño social": [
        2.54, 2.17, 2.21, 1.69, 1.65, 1.50, 1.42, 1.50, 1.52, 1.32, 1.30, 1.09, 0.97
    ]
}

fig = px.scatter_3d(data, x="Daño físico", y="Dependencia", z="Daño social", text="Sustancia")

# Cambiar el tamaño y el color de los puntos
fig.update_traces(marker=dict(size=10, color='#ffde59'))

# Cambiar el color de los puntos para sustancias legales (alcohol y tabaco) a gris
fig.update_traces(marker=dict(size=10), selector=dict(text=["Alcohol", "Tabaco"]), marker=dict(color='gray'))

# Ocultar el nombre de las sustancias
fig.update_traces(text='')

fig.update_layout(scene=dict(xaxis_title="Daño físico", yaxis_title="Dependencia", zaxis_title="Daño social"))
fig.show()
```

```python
import plotly.express as px
import pandas as pd

data = {
    "Sustancia": [
        "Heroína", "Cocaína", "Alcohol", "Ketamina", "Benzodiacepinas",
        "Anfetaminas", "Tabaco", "Marihuana", "Solventes", "LSD", "GHB", "Éxtasis", "Poppers"
    ],
    "Daño físico": [
        2.78, 2.33, 1.40, 2.00, 1.63, 1.81, 1.24, 0.99, 1.28, 1.13, 0.86, 1.05, 0.93
    ],
    "Dependencia": [
        3.00, 2.39, 1.93, 1.54, 1.83, 1.67, 2.21, 1.51, 1.01, 1.23, 1.19, 1.13, 0.87
    ],
    "Daño social": [
        2.54, 2.17, 2.21, 1.69, 1.65, 1.50, 1.42, 1.50, 1.52, 1.32, 1.30, 1.09, 0.97
    ]
}

df = pd.DataFrame(data)

# Actualizar el color de los puntos para sustancias legales (alcohol y tabaco) a gris
df['Color'] = df['Sustancia'].apply(lambda x: 'gray' if x in ['Alcohol', 'Tabaco'] else '#ffde59')

fig = px.scatter_3d(df, x="Daño físico", y="Dependencia", z="Daño social", text="Sustancia", color="Color")

# Ocultar el nombre de las sustancias
fig.update_traces(text='')

fig.update_layout(scene=dict(xaxis_title="Daño físico", yaxis_title="Dependencia", zaxis_title="Daño social"))
fig.show()
```

```python
import plotly.express as px
import pandas as pd

data = {
    "Sustancia": [
        "Heroína", "Cocaína", "Alcohol", "Ketamina", "Benzodiacepinas",
        "Anfetaminas", "Tabaco", "Marihuana", "Solventes", "LSD", "GHB", "Éxtasis", "Poppers"
    ],
    "Daño físico": [
        2.78, 2.33, 1.40, 2.00, 1.63, 1.81, 1.24, 0.99, 1.28, 1.13, 0.86, 1.05, 0.93
    ],
    "Dependencia": [
        3.00, 2.39, 1.93, 1.54, 1.83, 1.67, 2.21, 1.51, 1.01, 1.23, 1.19, 1.13, 0.87
    ],
    "Daño social": [
        2.54, 2.17, 2.21, 1.69, 1.65, 1.50, 1.42, 1.50, 1.52, 1.32, 1.30, 1.09, 0.97
    ]
}

# Actualizar el color de los puntos para sustancias legales (alcohol y tabaco) a gris
df['Color'] = df['Sustancia'].apply(lambda x: 'Legales' if x in ['Alcohol', 'Tabaco'] else 'Ilegales')

fig = px.scatter_3d(df, x="Daño físico", y="Dependencia", z="Daño social", text="Sustancia", color="Color",
                     color_discrete_map={"Legales": "red", "Ilegales": "purple"})

# Cambiar el tamaño de la figura y hacerla cuadrada
fig.update_layout(
    width=800,  # Ancho
    height=800,  # Alto
    scene=dict(xaxis_title="Daño físico", yaxis_title="Dependencia", zaxis_title="Daño social")
)

# Personalizar la leyenda
fig.update_traces(marker=dict(size=10))
fig.update_layout(legend_title_text="Leyenda")
fig.show()
```
